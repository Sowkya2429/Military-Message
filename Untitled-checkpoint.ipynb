```python
# Type your message :
# ×××××××××××××××××
# The encoded part is :
# The decoded part is :
# ×××××××××××××××××
from tkinter import *
from PIL import Image, ImageTk
from pygame import mixer

mixer.init() #Initialzing pyamge mixer

mixer.music.load('e.mp3') #Loading Music File

def play():
    mixer.music.play() #Playing Music with Pygame

def stop():
    mixer.music.stop()

window = Tk()

window.option_add('*Font', '40') 
window.minsize(1200,590)
window.title("MILITARY MESSAGE")
text = Text(window)

# Load image
i = Image.open("c.jpeg")
img=ImageTk.PhotoImage(i)

# Create label with image
label = Label(window,image=img)
label.place(x=0,y=0)
    

play_button = Button(window, text="Play Song",bg="green",relief=GROOVE, command=play).grid(row=3,column=10, padx = 50)
stop_buttom= Button(window, text="stop Song",bg="red",relief=GROOVE, command=stop).grid(row=4,column=10,padx = 50)

l1 = Label(window, text="Type your message:", fg="blue4")
l2 = Label(window, text="****")
l3 = Label(window, text="The encoded part is:", fg="green")
l5 = Label(window, text="****")
l6 = Label(window, text="Type your message:", fg="blue4")
l7 = Label(window, text="****")
l8 = Label(window, text="The decoded part is:",fg="green")
l9 = Label(window, text="****")
l1.grid(row=2, column=0, sticky=W, pady=20, padx=100)
l2.grid(row=3, column=0, sticky=W, pady=20, padx=100)
l3.grid(row=4, column=0, sticky=W, pady=20, padx=100)
l5.grid(row=6, column=0, sticky=W, pady=20, padx=100)
l6.grid(row=7, column=0, sticky=W, pady=20, padx=100)
l7.grid(row=8, column=0, sticky=W, pady=20, padx=100)
l8.grid(row=9, column=0, sticky=W, pady=20, padx=100)
l9.grid(row=10, column=0, sticky=W, pady=20, padx=100)



def encode():
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u',
                'v', 'w', 'x', 'y', 'z', ' ']

    index = ['j', 'x', 'r', 'i', 'd', 'z', 't', 'a', 'q', 'k', 'e', 'b', 'h', 'p', 'w', 'c', 'o', 'l', 'f', 's', 'v',
             'm', 'g', 'y', 'u', 'n', ' ']
    cipher_text = ""
    plain_text = e1.get()
    inc=5
    j=0
    i=0
    a=[0]*256
    duplicate=""
    
    for i in range(len(plain_text)):
        a[ord(plain_text[i])] = 0
        
    for letter in plain_text:
        position = alphabet.index(letter)
        x=letter in duplicate
        if (x):
            a[ord(letter)]=a[ord(letter)]+1
            m=a[ord(letter)]
            position=(position+inc*m)%26
            new_letter = index[position]
            cipher_text += new_letter.upper()
        else:  
            new_letter = index[position]
            cipher_text += new_letter.upper()
        if letter not in duplicate:
            duplicate+=letter.upper()
            
    encodetext = Label(window, text=cipher_text, fg="blue")
    encodetext.grid(row=4,column=1)

    

e1 = Entry(window)
e1.grid(row=2, column=1, pady=2)

Button(window, text="encode", bg="red", command=encode).grid(row=2, column=4)



def decode():
    index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u',
                'v', 'w', 'x', 'y', 'z', ' ']

    alphabet = ['j', 'x', 'r', 'i', 'd', 'z', 't', 'a', 'q', 'k', 'e', 'b', 'h', 'p', 'w', 'c', 'o', 'l', 'f', 's', 'v',
             'm',
             'g', 'y', 'u', 'n', ' ']
    cipher_text = e3.get()
    plain_text = ""
    for letter in cipher_text:
        position = alphabet.index(letter)
        plain_letter = index[position]
        plain_text += plain_letter.upper()
    encodetext = Label(window, text=plain_text, fg="blue")
    encodetext.grid(row=9,column=1)
    
 
e3 = Entry(window)
# this will arrange entry widgets
e3.grid(row=7, column=1, pady=2)
# e2.grid(row=4, column=1, pady=2)

Button(window, text="decode", bg="red", command=decode).grid(row=7, column=4)


window.mainloop()
```
